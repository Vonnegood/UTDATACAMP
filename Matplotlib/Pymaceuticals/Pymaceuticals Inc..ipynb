```python
trials = mouse_raw.merge(clinical_raw, on ="Mouse ID", how="inner")
trials.head()
# trials.count()
```

### Tumor Response to Treatment

```python
drug_time = trials.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
drug_time = pd.DataFrame(drug_time)
drug_time
```

```python
timebydrug = pd.pivot_table(drug_time, index=drug_time.index.get_level_values("Timepoint"), 
                            columns=drug_time.index.get_level_values("Drug"))
timebydrug
```
